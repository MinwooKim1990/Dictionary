# Recursive Algorithm

```python
nums=[1,2,3]
def permute(nums):
    if len(nums) == 1:
        return [nums]
    result = []
    for i in range(len(nums)):
        others = nums[:i] + nums[i+1:]
        other_permutations = permute(others)
        for permutation in other_permutations:
            result.append([nums[i]] + permutation)
    return result
print(permute(nums))
```

# Binary Search Algorithm


```python
def binary_search(arr, x):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == x:
            return mid
        elif arr[mid] < x:
            low = mid + 1
        else:
            high = mid - 1
    return -1
arr = [1, 2, 3, 4, 5, 6, 7, 8, 9]
x = 6
result = binary_search(arr, x)
if result != -1:
    print("Element is present at index", str(result))
else:
    print("Element is not present in the array")
```

# Sorting algorithms
These algorithms arrange elements of a list in a particular order, for example, bubble sort, insertion sort, merge sort, quick sort, etc.

```python
import time
import random
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr)//2
        L = arr[:mid]
        R = arr[mid:]
        merge_sort(L)
        merge_sort(R)
        i = j = k = 0
        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1
        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1
        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1
    return arr
def quick_sort(arr, low, high):
    if low < high:
        pi = partition(arr, low, high)
        quick_sort(arr, low, pi-1)
        quick_sort(arr, pi+1, high)
    return arr
def partition(arr, low, high):
    pivot = arr[high]
    i = low - 1
    for j in range(low, high):
        if arr[j] <= pivot:
            i = i + 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1
if __name__ == "__main__":
    arr = random.sample(range(1, 10001), 1000)
    start = time.time()
    bubble_sort(arr.copy())
    print("Bubble sort: ", time.time() - start)
    start = time.time()
    insertion_sort(arr.copy())
    print("Insertion sort: ", time.time() - start)
    start = time.time()
    merge_sort(arr.copy())
    print("Merge sort: ", time.time() - start)
    start = time.time()
    quick_sort(arr.copy(), 0, len(arr) - 1)
    print("Quick sort: ", time.time() - start)
```

# Dynamic Programming
Dynamic programming is a technique used to solve problems by breaking them down into simpler subproblems and storing the results of these subproblems to avoid redundant computations. Here's an advanced example using dynamic programming to solve the 0/1 knapsack problem, which is a classic optimization problem.

The 0/1 knapsack problem is defined as follows:
Given a set of items, each with a weight and a value, determine the items to include in a knapsack such that the total weight is less than or equal to a given limit and the total value is as large as possible.

```python
def knapSack(W, weights, values, n):
    # Create a memoization table to store the optimal solution for each subproblem
    dp = [[0 for w in range(W + 1)] for i in range(n + 1)]

    # Build the table in a bottom-up manner
    for i in range(n + 1):
        for w in range(W + 1):
            if i == 0 or w == 0:
                dp[i][w] = 0
            elif weights[i-1] <= w:
                dp[i][w] = max(values[i-1] + dp[i-1][w-weights[i-1]], dp[i-1][w])
            else:
                dp[i][w] = dp[i-1][w]

    # Now backtrack to find the items that were included in the optimal solution
    items_included = []
    i, w = n, W
    while i > 0 and w > 0:
        if dp[i][w] != dp[i-1][w]:
            items_included.append(i-1)
            w -= weights[i-1]
        i -= 1

    # Return the optimal value and the items included
    return dp[n][W], items_included

if __name__ == "__main__":
    values = [60, 100, 120]
    weights = [10, 20, 30]
    W = 50
    n = len(values)

    max_value, items = knapSack(W, weights, values, n)
    print("The maximum value that can be obtained is:", max_value)
    print("Items included (indexed from 0):", items)
```

# Graph algorithms
Algorithms for working with graphs, such as depth-first search, breadth-first search, shortest path algorithms, and minimum spanning tree algorithms.
Depth-First Search (DFS), Breadth-First Search (BFS), Dijkstra's shortest path algorithm, and Kruskal's minimum spanning tree algorithm using an adjacency list representation for the graph

```python
from collections import defaultdict, deque
from heapq import heappop, heappush
from functools import total_ordering

class Graph:
    def __init__(self, vertices, directed=False):
        self.graph = defaultdict(list)
        self.directed = directed
        self.vertices = vertices

    def add_edge(self, u, v, weight):
        self.graph[u].append((v, weight))
        if not self.directed:
            self.graph[v].append((u, weight))

    def dfs(self, start):
        visited = set()
        stack = [start]

        while stack:
            vertex = stack.pop()
            if vertex not in visited:
                visited.add(vertex)
                stack.extend(neighbor for neighbor, _ in self.graph[vertex] if neighbor not in visited)

        return visited

    def bfs(self, start):
        visited = set()
        queue = deque([start])

        while queue:
            vertex = queue.popleft()
            if vertex not in visited:
                visited.add(vertex)
                queue.extend(neighbor for neighbor, _ in self.graph[vertex] if neighbor not in visited)

        return visited

    def dijkstra(self, start, end):
        heap = [(0, start)]
        visited = set()
        while heap:
            (cost, current) = heappop(heap)
            if current in visited:
                continue
            visited.add(current)
            if current == end:
                return cost
            for neighbor, weight in self.graph[current]:
                if neighbor not in visited:
                    heappush(heap, (cost + weight, neighbor))

        return -1

    def kruskal(self):
        edges = [(u, v, weight) for u in self.graph for v, weight in self.graph[u]]
        edges.sort(key=lambda x: x[2])

        parent = {i: i for i in range(self.vertices)}
        rank = {i: 0 for i in range(self.vertices)}

        def find(vertex):
            if parent[vertex] != vertex:
                parent[vertex] = find(parent[vertex])
            return parent[vertex]

        def union(u_root, v_root):
            if rank[u_root] > rank[v_root]:
                parent[v_root] = u_root
            else:
                parent[u_root] = v_root
                if rank[u_root] == rank[v_root]:
                    rank[v_root] += 1

        minimum_spanning_tree = []
        for u, v, weight in edges:
            u_root, v_root = find(u), find(v)
            if u_root != v_root:
                minimum_spanning_tree.append((u, v, weight))
                union(u_root, v_root)

        return minimum_spanning_tree

if __name__ == "__main__":
    g = Graph(9)
    edges = [
        (0, 1, 4), (0, 7, 8), (1, 7, 11), (1, 2, 8), (2, 8, 2),
        (2, 3, 7), (2, 5, 4), (3, 5, 14), (3, 4, 9), (4, 5, 10),
        (5, 6, 2), (6, 8, 6), (6, 7, 1), (7, 8, 7)
    ]

    for edge in edges:
        g.add_edge(*edge)

    # DFS
    visited_dfs = g.dfs(0)
    print("DFS traversal:", visited_dfs)

    # BFS
    visited_bfs = g.bfs(0)
    print("BFS traversal:", visited_bfs)

    # Dijkstra's Shortest Path
    shortest_path = g.dijkstra(0, 4)
    print("Shortest path from 0 to 4 using Dijkstra's algorithm:", shortest_path)

    # Kruskal's Minimum Spanning Tree
    min_spanning_tree = g.kruskal()
    print("Minimum spanning tree using Kruskal's algorithm:", min_spanning_tree)
```

# Greedy algorithm

Algorithms that make the locally optimal choice at each stage in the hope of finding a global optimum.
Given a set of items, each with a weight and a value, and a knapsack with a maximum weight capacity, determine the maximum value that can be obtained by including fractions of items, such that the total weight does not exceed the knapsack capacity.

```python
def fractional_knapsack(items, capacity):
    # Sort items by value-to-weight ratio in descending order
    items.sort(key=lambda x: x['value'] / x['weight'], reverse=True)

    total_value = 0
    for item in items:
        if capacity == 0:
            break

        weight_to_take = min(item['weight'], capacity)
        fraction = weight_to_take / item['weight']
        total_value += item['value'] * fraction
        capacity -= weight_to_take

    return total_value

if __name__ == "__main__":
    items = [
        {'value': 60, 'weight': 10},
        {'value': 100, 'weight': 20},
        {'value': 120, 'weight': 30},
    ]
    capacity = 50

    max_value = fractional_knapsack(items, capacity)
    print("The maximum value that can be obtained is:", max_value)
```

# Backtracking algorithm
Backtracking: A general algorithmic technique that incrementally builds candidates to the solutions, and abandons a candidate ("backtracks") as soon as it determines that the candidate cannot possibly be completed to a valid solution.

Place N chess queens on an N x N chessboard so that no two queens threaten each other (i.e., they should not be in the same row, column, or diagonal).

```python
def is_safe(board, row, col, n):
    # Check the same column
    for i in range(row):
        if board[i][col] == 1:
            return False

    # Check upper diagonal on the left side
    for i, j in zip(range(row, -1, -1), range(col, -1, -1)):
        if board[i][j] == 1:
            return False

    # Check upper diagonal on the right side
    for i, j in zip(range(row, -1, -1), range(col, n)):
        if board[i][j] == 1:
            return False

    return True

def n_queens_util(board, row, n):
    if row == n:
        return True

    for col in range(n):
        if is_safe(board, row, col, n):
            board[row][col] = 1

            if n_queens_util(board, row + 1, n):
                return True

            board[row][col] = 0

    return False

def n_queens(n):
    board = [[0 for _ in range(n)] for _ in range(n)]

    if not n_queens_util(board, 0, n):
        return None

    return board

if __name__ == "__main__":
    n = 8
    result = n_queens(n)

    if result is not None:
        for row in result:
            print(row)
    else:
        print("No solution found.")
```

# Randomize algorithm
Randomized algorithms: Algorithms that use randomness as part of their logic.
Reservoir Sampling algorithm, which is used to select k items randomly from a list of items of unknown size. This algorithm is particularly useful when the size of the list is too large or expensive to bring into memory.

```python
import random

def reservoir_sampling(stream, k):
    reservoir = []

    for i, item in enumerate(stream):
        if i < k:
            reservoir.append(item)
        else:
            random_index = random.randint(0, i)
            if random_index < k:
                reservoir[random_index] = item

    return reservoir

def generate_stream(n):
    for i in range(n):
        yield i

if __name__ == "__main__":
    k = 10
    n = 100

    stream = generate_stream(n)
    sampled_items = reservoir_sampling(stream, k)
    print("Sampled items:", sampled_items)
```

# Random Walk

```python
import matplotlib.pyplot as plt
import numpy as np
num_walks = 3 # Set up the parameters
num_steps = 500
step_size = 1
start_positions = [[0, 0, 0], [1, 1, 1], [2, -1, 3]] # Define the starting positions for each walker
walks = [] # Generate the random walks
for i in range(num_walks):
    x, y, z = start_positions[i]
    walk = [[x], [y], [z]]
    for j in range(num_steps):
        step = step_size * np.random.normal(0, 4, 1).astype(int)[0]
        xyz=np.random.randint(0,3)
        if xyz == 0:
            x+=step
        elif xyz == 1:
            y+=step
        else :
            z+=step
        walk[0].append(x)
        walk[1].append(y)
        walk[2].append(z)
    walks.append(walk)
fig = plt.figure(figsize=(8, 8)) # Plot the random walks
ax = fig.add_subplot(111, projection='3d')
for i in range(num_walks):
    x, y, z = walks[i]
    ax.plot(x, y, z)
plt.show()
```

# Black-Scholes Equation

```python
import numpy as np
from scipy.stats import norm
def black_scholes(S, K, T, r, sigma, option_type):
    d1 = (np.log(S/K) + (r + 0.5*sigma**2)*T) / (sigma*np.sqrt(T))
    d2 = d1 - sigma*np.sqrt(T)
    if option_type == 'call':
        price = S*norm.cdf(d1) - K*np.exp(-r*T)*norm.cdf(d2)
    elif option_type == 'put':
        price = K*np.exp(-r*T)*norm.cdf(-d2) - S*norm.cdf(-d1)
    return price
S = 100   # current stock price
K = 110   # strike price
T = 1     # time to maturity (in years)
r = 0.05  # risk-free interest rate
sigma = 0.2  # volatility
option_type = 'call'  # type of option ('call' or 'put')
price = black_scholes(S, K, T, r, sigma, option_type)
print(price)  # print the option price
```

```python
import cv2
import numpy as np
from numba import jit, njit
import os
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import timeit
os.environ['NUMBA_WARNINGS'] = '0'
@jit(nopython=True)
#@njit(parallel = False)
def computedisparity2(left, right, deep, window):
    max_disparity = deep
    size = left.shape
    kernel_size = window
    disp = np.zeros(size)    
    for y in range(size[0]):    
        y_kernel = (max(0, y-int(kernel_size/2)), min(size[0]-1, y+int(kernel_size/2)))
        for x in range(size[1]):
            x_kernel = (max(0, x-int(kernel_size/2)), min(size[1]-1, x+int(kernel_size/2)))
            max_disp = min([max_disparity, size[1]-x_kernel[1]])
            template = right[y_kernel[0]:y_kernel[1], x_kernel[0]:x_kernel[1]]
            SSD = np.zeros(max_disp + 1)
            for i in range(max_disp+1):
                block = left[y_kernel[0]:y_kernel[1], x_kernel[0]+i:x_kernel[1]+i]
                SSD[i] = np.sum((template-block)**2)
            disparity = np.amin(SSD)
            dis_index = np.argmin(SSD)
            if ((dis_index == 0) or (dis_index == max_disp)):
                disp[y, x] = dis_index
            else:
                C3 = SSD[dis_index - 1]
                C2 = SSD[dis_index]
                C1 = SSD[dis_index + 1]
                disp[y, x] = dis_index - (0.5 * (C3 - C1) / (C1 + C3 - (2*C2)))
    return disp
left = cv2.imread('img/left.png')
left = cv2.cvtColor(left, cv2.COLOR_BGR2GRAY)
right = cv2.imread('img/right.png')
right = cv2.cvtColor(right, cv2.COLOR_BGR2GRAY)
_, image_cells = plt.subplots(1, 2, figsize=(20, 20))
image_cells[0].imshow(left)
image_cells[0].set_title('left image')
image_cells[1].imshow(right)
image_cells[1].set_title('right image')
plt.show()
start = timeit.default_timer()
stereo = computedisparity2(left, right, 50, 20)
stop = timeit.default_timer()
print('Time: ', stop - start)  
plt.imshow(stereo, cmap='gray')
z = stereo
x = np.arange(stereo.shape[1])
y = np.arange(stereo.shape[0])
X, Y = np.meshgrid(x, y)
fig = plt.figure(figsize = (12, 7))
ax = plt.axes(projection='3d')
ax.plot_surface(-X, Y, z, cmap='viridis', alpha=0.7)
ax.view_init(80, 90, 0)
ax.set_title('3D reconstruction')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('stereo')
plt.show()
```
