# Monte-Carlo Integration

```python
import numpy as np
import matplotlib.pyplot as plt
def func(x): # Define the function to be integrated
    return x**2
a, b, N= 0, 2, 100000 # Limit 0~2 and number of random points 100000 
c = b**2
x = np.random.uniform(a, b, N) # Generate random x and y values between the limits of integration
y = np.random.uniform(a, c, N)
f = func(x) # Evaluate the function at the random x values
below = np.sum(y < f) # Count the number of points below the function
integral = b*c * (below/N) # Calculate the area under the function
print("Monte Carlo Estimate of the Integral: ", integral, 'Analytic Integral:', 1/3*b**3) # Print the result
plt.scatter(x, y, alpha=0.1, s=3, c='g', label='Random Points') # Plot the results
plt.scatter(x, f, c='b', label='Function')
plt.fill_between(np.linspace(a,b,num=20), 0, func(np.linspace(a,b,num=20)), alpha=0.8, color='r', label='Area Under Function')
plt.legend()
plt.xlim(0,b)
plt.ylim(0,c)
plt.show()
```

# Binary Search Algorithm


```python
def binary_search(arr, x):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == x:
            return mid
        elif arr[mid] < x:
            low = mid + 1
        else:
            high = mid - 1
    return -1
arr = [1, 2, 3, 4, 5, 6, 7, 8, 9]
x = 6
result = binary_search(arr, x)
if result != -1:
    print("Element is present at index", str(result))
else:
    print("Element is not present in the array")
```
