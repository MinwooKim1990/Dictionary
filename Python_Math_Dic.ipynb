# Monte-Carlo Integration

```python
import numpy as np
import matplotlib.pyplot as plt
def func(x): # Define the function to be integrated
    return x**2
a, b, N= 0, 2, 100000 # Limit 0~2 and number of random points 100000 
c = b**2
x = np.random.uniform(a, b, N) # Generate random x and y values between the limits of integration
y = np.random.uniform(a, c, N)
f = func(x) # Evaluate the function at the random x values
below = np.sum(y < f) # Count the number of points below the function
integral = b*c * (below/N) # Calculate the area under the function
print("Monte Carlo Estimate of the Integral: ", integral, 'Analytic Integral:', 1/3*b**3) # Print the result
plt.scatter(x, y, alpha=0.1, s=3, c='g', label='Random Points') # Plot the results
plt.scatter(x, f, c='b', label='Function')
plt.fill_between(np.linspace(a,b,num=20), 0, func(np.linspace(a,b,num=20)), alpha=0.8, color='r', label='Area Under Function')
plt.legend()
plt.xlim(0,b)
plt.ylim(0,c)
plt.show()
```

# Recursive Algorithm

```python
nums=[1,2,3]
def permute(nums):
    if len(nums) == 1:
        return [nums]
    result = []
    for i in range(len(nums)):
        others = nums[:i] + nums[i+1:]
        other_permutations = permute(others)
        for permutation in other_permutations:
            result.append([nums[i]] + permutation)
    return result
print(permute(nums))
```

# Binary Search Algorithm


```python
def binary_search(arr, x):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == x:
            return mid
        elif arr[mid] < x:
            low = mid + 1
        else:
            high = mid - 1
    return -1
arr = [1, 2, 3, 4, 5, 6, 7, 8, 9]
x = 6
result = binary_search(arr, x)
if result != -1:
    print("Element is present at index", str(result))
else:
    print("Element is not present in the array")
```

Sorting algorithms: These algorithms arrange elements of a list in a particular order, for example, bubble sort, insertion sort, merge sort, quick sort, etc.
Dynamic programming: A method for solving complex problems by breaking them down into smaller, simpler subproblems and solving each subproblem only once.
Graph algorithms: Algorithms for working with graphs, such as depth-first search, breadth-first search, shortest path algorithms, and minimum spanning tree algorithms.
Divide and conquer: A problem-solving strategy that involves breaking down a problem into smaller subproblems that are easier to solve, then combining the solutions of the subproblems to solve the original problem.
Greedy algorithms: Algorithms that make the locally optimal choice at each stage in the hope of finding a global optimum.
Backtracking: A general algorithmic technique that incrementally builds candidates to the solutions, and abandons a candidate ("backtracks") as soon as it determines that the candidate cannot possibly be completed to a valid solution.
Randomized algorithms: Algorithms that use randomness as part of their logic.

```python
import time
import random
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr)//2
        L = arr[:mid]
        R = arr[mid:]
        merge_sort(L)
        merge_sort(R)
        i = j = k = 0
        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1
        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1
        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1
    return arr
def quick_sort(arr, low, high):
    if low < high:
        pi = partition(arr, low, high)
        quick_sort(arr, low, pi-1)
        quick_sort(arr, pi+1, high)
    return arr
def partition(arr, low, high):
    pivot = arr[high]
    i = low - 1
    for j in range(low, high):
        if arr[j] <= pivot:
            i = i + 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1
arr = random.sample(range(1, 10001), 1000)
start = time.time()
bubble_sort(arr.copy())
print("Bubble sort: ", time.time() - start)
start = time.time()
insertion_sort(arr.copy())
print("Insertion sort: ", time.time() - start)
start = time.time()
merge_sort(arr.copy())
print("Merge sort: ", time.time() - start)
start = time.time()
quick_sort(arr.copy(), 0, len(arr) - 1)
print("Quick sort: ", time.time() - start)
```

# Singular Value Decomposition (SVD)

```python
import numpy as np

# Create a random matrix with 4 rows and 3 columns
A = np.random.rand(4, 3)

# Perform singular value decomposition on the matrix A
U, s, Vt = np.linalg.svd(A)

# Reconstruct the original matrix from the SVD components
Sigma = np.zeros_like(A)
Sigma[:min(A.shape), :min(A.shape)] = np.diag(s)
B = np.dot(U, np.dot(Sigma, Vt))

# Print the original matrix A, reconstructed matrix B, and SVD components
print("Matrix A:\n", A)
print("\nReconstructed matrix B:\n", B)
print("\nLeft singular vectors (U):\n", U)
print("\nSingular values (s):\n", s)
print("\nRight singular vectors (Vt):\n", Vt)
```

```python
import numpy as np
import matplotlib.pyplot as plt
class DEq_Solver:
    def solve(self,x0,t0,t1,delta_t,terminateCondition=None,circular_index=None):
        self.initialise(x0,t0,t1,delta_t)
        while self.t <= self.t1:
            self.record(self.t,self.x)
            if terminateCondition is not None:
                if terminateCondition(self.x):
                    break
            self.makeStep()
        return self.history
    def initialise(self,x0,t0,t1,delta_t):
        self.x       = x0.copy()
        self.t       = t0
        self.t1      = t1
        self.delta_t = delta_t
        self.history = []
    def record(self,t,x):
        self.history.append(  ( t, x.copy() )  )
    def timeSteps(self):
        ts,vs=list(zip(*self.history))
        return np.array(ts).T
    def coordinateSteps(self):
        ts,vs=list(zip(*self.history))
        return np.array(vs).T
    def makeStep(self):
        raise NotImplemented
class RK2Solver(DEq_Solver):
    def __init__(self,kernel):
        self.kernel  = kernel
    def makeStep(self):
        self.x=self.x+self.kernel.dx_dt(self.x+self.delta_t*self.kernel.dx_dt(self.x,self.t)/2,self.t+self.delta_t/2)*self.delta_t
        self.t=self.t+self.delta_t 
class RK4Solver(DEq_Solver):
    def __init__(self,kernel):
        self.kernel  = kernel
    def makeStep(self):
        x1=self.x
        t1=self.t
        x2=self.x+self.kernel.dx_dt(x1,t1)*self.delta_t/2
        t2=self.t+self.delta_t/2
        x3=self.x+self.kernel.dx_dt(x2,t2)*self.delta_t/2
        t3=self.t+self.delta_t/2
        x4=self.x+self.kernel.dx_dt(x3,t3)*self.delta_t
        t4=self.t+self.delta_t
        self.x=self.x+self.delta_t/6*(self.kernel.dx_dt(x1,t1)+2*self.kernel.dx_dt(x2,t2)+2*self.kernel.dx_dt(x3,t3)+self.kernel.dx_dt(x4,t4))
        self.t=self.t+self.delta_t
class EulersMethod(DEq_Solver):
    def __init__(self,kernel):
        self.kernel  = kernel
    def makeStep(self):
        self.x=self.x+self.kernel.dx_dt(self.x,self.t)*self.delta_t
        self.t=self.t+self.delta_t
class DifferentialEquation:
    def dx_dt(self,x,t):
        raise NotImplemented
class TestFunction(DifferentialEquation):
    def dx_dt(self,x,t):
        return np.sin(t)*np.cos(x)
t0 = 0
t1 = 10
delta_t = 0.1
x0 = np.array([0.5])
solver_euler = EulersMethod(TestFunction())
solver_rk2 = RK2Solver(TestFunction())
solver_rk4 = RK4Solver(TestFunction())
t = np.linspace(t0, t1, int((t1 - t0) / delta_t) + 1)
ground_truth = np.sin(t) + 0.5
class SimpleHarmonicOscillator: # define the differential equation
    def dx_dt(self, x, t):
        k = 1
        m = 1
        return np.array([x[1], -k/m*x[0]])
# set up the initial conditions and time interval
x0 = np.array([1, 0]) # initial position and velocity
t0 = 0 # start time
t1 = 10 # end time
delta_t = 0.1 # time step
def ground_truth(t): #define the ground truth function
    return np.array([np.cos(t), -np.sin(t)])
SHO = SimpleHarmonicOscillator() # create the solver objects
euler_cromer = EulersMethod(SHO)
rk2 = RK2Solver(SHO)
rk4 = RK4Solver(SHO)
euler_cromer_soln = euler_cromer.solve(x0, t0, t1, delta_t) #solve the differential equation using each solver
rk2_soln = rk2.solve(x0, t0, t1, delta_t)
rk4_soln = rk4.solve(x0, t0, t1, delta_t)
fig, ax = plt.subplots() #plot the results
t = np.linspace(t0, t1, int((t1-t0)/delta_t+1)) 
ax.plot(t, ground_truth(t)[0], label='Ground Truth') #plot ground truth
ax.scatter(euler_cromer.timeSteps(), euler_cromer.coordinateSteps()[0], c='orange', s=10, label='Euler-Cromer') #plot Euler-Cromer
ax.scatter(rk2.timeSteps(), rk2.coordinateSteps()[0], c='r', s=10, label='RK2') #plot RK2
ax.scatter(rk4.timeSteps(), rk4.coordinateSteps()[0], c='g', s=10, label='RK4') #plot RK4
ax.set_xlabel('Time')
ax.set_ylabel('Position')
ax.legend()
plt.show()
```

# Random Walk

```python
import matplotlib.pyplot as plt
import numpy as np
num_walks = 3 # Set up the parameters
num_steps = 500
step_size = 1
start_positions = [[0, 0, 0], [1, 1, 1], [2, -1, 3]] # Define the starting positions for each walker
walks = [] # Generate the random walks
for i in range(num_walks):
    x, y, z = start_positions[i]
    walk = [[x], [y], [z]]
    for j in range(num_steps):
        step = step_size * np.random.normal(0, 4, 1).astype(int)[0]
        xyz=np.random.randint(0,3)
        if xyz == 0:
            x+=step
        elif xyz == 1:
            y+=step
        else :
            z+=step
        walk[0].append(x)
        walk[1].append(y)
        walk[2].append(z)
    walks.append(walk)
fig = plt.figure(figsize=(8, 8)) # Plot the random walks
ax = fig.add_subplot(111, projection='3d')
for i in range(num_walks):
    x, y, z = walks[i]
    ax.plot(x, y, z)
plt.show()
```

# Black-Scholes Equation

```python
import numpy as np
from scipy.stats import norm
def black_scholes(S, K, T, r, sigma, option_type):
    d1 = (np.log(S/K) + (r + 0.5*sigma**2)*T) / (sigma*np.sqrt(T))
    d2 = d1 - sigma*np.sqrt(T)
    if option_type == 'call':
        price = S*norm.cdf(d1) - K*np.exp(-r*T)*norm.cdf(d2)
    elif option_type == 'put':
        price = K*np.exp(-r*T)*norm.cdf(-d2) - S*norm.cdf(-d1)
    return price
S = 100   # current stock price
K = 110   # strike price
T = 1     # time to maturity (in years)
r = 0.05  # risk-free interest rate
sigma = 0.2  # volatility
option_type = 'call'  # type of option ('call' or 'put')
price = black_scholes(S, K, T, r, sigma, option_type)
print(price)  # print the option price
```

```python
import cv2
import numpy as np
from numba import jit, njit
import os
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import timeit
os.environ['NUMBA_WARNINGS'] = '0'
@jit(nopython=True)
#@njit(parallel = False)
def computedisparity2(left, right, deep, window):
    max_disparity = deep
    size = left.shape
    kernel_size = window
    disp = np.zeros(size)    
    for y in range(size[0]):    
        y_kernel = (max(0, y-int(kernel_size/2)), min(size[0]-1, y+int(kernel_size/2)))
        for x in range(size[1]):
            x_kernel = (max(0, x-int(kernel_size/2)), min(size[1]-1, x+int(kernel_size/2)))
            max_disp = min([max_disparity, size[1]-x_kernel[1]])
            template = right[y_kernel[0]:y_kernel[1], x_kernel[0]:x_kernel[1]]
            SSD = np.zeros(max_disp + 1)
            for i in range(max_disp+1):
                block = left[y_kernel[0]:y_kernel[1], x_kernel[0]+i:x_kernel[1]+i]
                SSD[i] = np.sum((template-block)**2)
            disparity = np.amin(SSD)
            dis_index = np.argmin(SSD)
            if ((dis_index == 0) or (dis_index == max_disp)):
                disp[y, x] = dis_index
            else:
                C3 = SSD[dis_index - 1]
                C2 = SSD[dis_index]
                C1 = SSD[dis_index + 1]
                disp[y, x] = dis_index - (0.5 * (C3 - C1) / (C1 + C3 - (2*C2)))
    return disp
left = cv2.imread('img/left.png')
left = cv2.cvtColor(left, cv2.COLOR_BGR2GRAY)
right = cv2.imread('img/right.png')
right = cv2.cvtColor(right, cv2.COLOR_BGR2GRAY)
start = timeit.default_timer()
stereo = computedisparity2(left, right, 50, 20)
stop = timeit.default_timer()
print('Time: ', stop - start)  
plt.imshow(stereo, cmap='gray')
z = stereo
x = np.arange(stereo.shape[1])
y = np.arange(stereo.shape[0])
X, Y = np.meshgrid(x, y)
fig = plt.figure(figsize = (12, 7))
ax = plt.axes(projection='3d')
ax.plot_surface(-X, Y, z, cmap='viridis', alpha=0.7)
ax.view_init(80, 90, 0)
ax.set_title('3D reconstruction')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('stereo')
plt.show()
```
