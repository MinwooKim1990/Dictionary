# ListNode
A ListNode is a basic building block of a singly-linked list data structure. It usually contains a value and a reference to the next node in the list.

```python
class ListNode: # Defining a ListNode Class
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
# Creating a linked list with three nodes
node1 = ListNode(1) # Initialize nodes
node2 = ListNode(2)
node3 = ListNode(3)

node1.next = node2 # Link nodes together
node2.next = node3
# Printing a linked list
def print_linked_list(head):
    current = head
    while current is not None:
        print(current.val, end=" -> ")
        current = current.next
    print("None")
print_linked_list(node1)
# Inserting a new node at the beginning of the list
new_head = ListNode(0)
new_head.next = node1
print_linked_list(new_head)
# Inserting a new node after a specific node
def insert_after_node(node, value):
    new_node = ListNode(value)
    new_node.next = node.next
    node.next = new_node
insert_after_node(node1, 1.5)
print_linked_list(new_head)
# Deleting a node with specific value
def delete_node(head, value):
    dummy = ListNode(0)
    dummy.next = head
    current = dummy
    while current.next is not None:
        if current.next.val == value:
            current.next = current.next.next
            break
        current = current.next
    return dummy.next
new_head = delete_node(new_head, 1.5)
print_linked_list(new_head)
```

# Recursive Algorithm

```python
nums=[1,2,3]
def permute(nums):
    if len(nums) == 1:
        return [nums]
    result = []
    for i in range(len(nums)):
        others = nums[:i] + nums[i+1:]
        other_permutations = permute(others)
        for permutation in other_permutations:
            result.append([nums[i]] + permutation)
    return result
print(permute(nums))
```

# Binary Search Algorithm


```python
def binary_search(arr, x):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == x:
            return mid
        elif arr[mid] < x:
            low = mid + 1
        else:
            high = mid - 1
    return -1
arr = [1, 2, 3, 4, 5, 6, 7, 8, 9]
x = 6
result = binary_search(arr, x)
if result != -1:
    print("Element is present at index", str(result))
else:
    print("Element is not present in the array")
```

A contiguous region is one where there is a connected group of 0's going in one or more of four directions: up, down, left, or right. For example: if strArr is ["10111", "10101", "11101", "11111"], then this looks like the following matrix:

```python
def SearchingChallenge(strArr):
    matrix = [[int(ch) for ch in row] for row in strArr] # string to int matrix
    num_rows = len(matrix) # size
    num_cols = len(matrix[0]) # size
    def dfs(i,j): # deep first search
        if i<0 or j<0 or i>= num_rows or j>=num_cols or matrix[i][j] !=0: # out when its not possible
            return
        matrix[i][j] = 1 # mark as 1 to erase previous 0
        # search up, down, left, right
        dfs(i-1,j) 
        dfs(i+1,j)
        dfs(i,j-1)
        dfs(i,j+1)
    num_holes =0 # initialise number of holes
    for i in range(num_rows):
        for j in range(num_cols): # search from 0,0
            if matrix[i][j]==0: # find 0
                num_holes +=1 # find hole
                dfs(i,j) # do deep first search
    return num_holes
input=["01111", "01101", "00011", "11110"] 
print(SearchingChallenge(input))
```

# Sorting algorithms
These algorithms arrange elements of a list in a particular order, for example, bubble sort, insertion sort, merge sort, quick sort, etc.

```python
import time
import random
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr)//2
        L = arr[:mid]
        R = arr[mid:]
        merge_sort(L)
        merge_sort(R)
        i = j = k = 0
        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1
        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1
        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1
    return arr
def quick_sort(arr, low, high):
    if low < high:
        pi = partition(arr, low, high)
        quick_sort(arr, low, pi-1)
        quick_sort(arr, pi+1, high)
    return arr
def partition(arr, low, high):
    pivot = arr[high]
    i = low - 1
    for j in range(low, high):
        if arr[j] <= pivot:
            i = i + 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1
if __name__ == "__main__":
    arr = random.sample(range(1, 10001), 1000)
    start = time.time()
    bubble_sort(arr.copy())
    print("Bubble sort: ", time.time() - start)
    start = time.time()
    insertion_sort(arr.copy())
    print("Insertion sort: ", time.time() - start)
    start = time.time()
    merge_sort(arr.copy())
    print("Merge sort: ", time.time() - start)
    start = time.time()
    quick_sort(arr.copy(), 0, len(arr) - 1)
    print("Quick sort: ", time.time() - start)
```

# Dynamic Programming
Dynamic programming is a technique used to solve problems by breaking them down into simpler subproblems and storing the results of these subproblems to avoid redundant computations. Here's an advanced example using dynamic programming to solve the 0/1 knapsack problem, which is a classic optimization problem.

The 0/1 knapsack problem is defined as follows:
Given a set of items, each with a weight and a value, determine the items to include in a knapsack such that the total weight is less than or equal to a given limit and the total value is as large as possible.

This code defines a function knapSack that solves the 0/1 knapsack problem using dynamic programming. It creates a memoization table dp to store the optimal solution for each subproblem and builds the table in a bottom-up manner. After constructing the table, it backtracks to find the items that were included in the optimal solution. The function returns the maximum value that can be obtained and the items included in the knapsack.

You can customize the values, weights, and W variables to test the code with different inputs.

```python
def knapSack(W, weights, values, n):
    # Create a memoization table to store the optimal solution for each subproblem
    dp = [[0 for w in range(W + 1)] for i in range(n + 1)]

    # Build the table in a bottom-up manner
    for i in range(n + 1):
        for w in range(W + 1):
            if i == 0 or w == 0:
                dp[i][w] = 0
            elif weights[i-1] <= w:
                dp[i][w] = max(values[i-1] + dp[i-1][w-weights[i-1]], dp[i-1][w])
            else:
                dp[i][w] = dp[i-1][w]

    # Now backtrack to find the items that were included in the optimal solution
    items_included = []
    i, w = n, W
    while i > 0 and w > 0:
        if dp[i][w] != dp[i-1][w]:
            items_included.append(i-1)
            w -= weights[i-1]
        i -= 1

    # Return the optimal value and the items included
    return dp[n][W], items_included

if __name__ == "__main__":
    values = [60, 100, 120]
    weights = [10, 20, 30]
    W = 50
    n = len(values)

    max_value, items = knapSack(W, weights, values, n)
    print("The maximum value that can be obtained is:", max_value)
    print("Items included (indexed from 0):", items)
```

# Graph algorithms
Algorithms for working with graphs, such as depth-first search, breadth-first search, shortest path algorithms, and minimum spanning tree algorithms.
Depth-First Search (DFS), Breadth-First Search (BFS), Dijkstra's shortest path algorithm, and Kruskal's minimum spanning tree algorithm using an adjacency list representation for the graph

```python
from collections import defaultdict, deque
from heapq import heappop, heappush
from functools import total_ordering

class Graph:
    def __init__(self, vertices, directed=False):
        self.graph = defaultdict(list)
        self.directed = directed
        self.vertices = vertices

    def add_edge(self, u, v, weight):
        self.graph[u].append((v, weight))
        if not self.directed:
            self.graph[v].append((u, weight))

    def dfs(self, start):
        visited = set()
        stack = [start]

        while stack:
            vertex = stack.pop()
            if vertex not in visited:
                visited.add(vertex)
                stack.extend(neighbor for neighbor, _ in self.graph[vertex] if neighbor not in visited)

        return visited

    def bfs(self, start):
        visited = set()
        queue = deque([start])

        while queue:
            vertex = queue.popleft()
            if vertex not in visited:
                visited.add(vertex)
                queue.extend(neighbor for neighbor, _ in self.graph[vertex] if neighbor not in visited)

        return visited

    def dijkstra(self, start, end):
        heap = [(0, start)]
        visited = set()
        while heap:
            (cost, current) = heappop(heap)
            if current in visited:
                continue
            visited.add(current)
            if current == end:
                return cost
            for neighbor, weight in self.graph[current]:
                if neighbor not in visited:
                    heappush(heap, (cost + weight, neighbor))

        return -1

    def kruskal(self):
        edges = [(u, v, weight) for u in self.graph for v, weight in self.graph[u]]
        edges.sort(key=lambda x: x[2])

        parent = {i: i for i in range(self.vertices)}
        rank = {i: 0 for i in range(self.vertices)}

        def find(vertex):
            if parent[vertex] != vertex:
                parent[vertex] = find(parent[vertex])
            return parent[vertex]

        def union(u_root, v_root):
            if rank[u_root] > rank[v_root]:
                parent[v_root] = u_root
            else:
                parent[u_root] = v_root
                if rank[u_root] == rank[v_root]:
                    rank[v_root] += 1

        minimum_spanning_tree = []
        for u, v, weight in edges:
            u_root, v_root = find(u), find(v)
            if u_root != v_root:
                minimum_spanning_tree.append((u, v, weight))
                union(u_root, v_root)

        return minimum_spanning_tree

if __name__ == "__main__":
    g = Graph(9)
    edges = [
        (0, 1, 4), (0, 7, 8), (1, 7, 11), (1, 2, 8), (2, 8, 2),
        (2, 3, 7), (2, 5, 4), (3, 5, 14), (3, 4, 9), (4, 5, 10),
        (5, 6, 2), (6, 8, 6), (6, 7, 1), (7, 8, 7)
    ]

    for edge in edges:
        g.add_edge(*edge)

    # DFS
    visited_dfs = g.dfs(0)
    print("DFS traversal:", visited_dfs)

    # BFS
    visited_bfs = g.bfs(0)
    print("BFS traversal:", visited_bfs)

    # Dijkstra's Shortest Path
    shortest_path = g.dijkstra(0, 4)
    print("Shortest path from 0 to 4 using Dijkstra's algorithm:", shortest_path)

    # Kruskal's Minimum Spanning Tree
    min_spanning_tree = g.kruskal()
    print("Minimum spanning tree using Kruskal's algorithm:", min_spanning_tree)
```

# Greedy algorithm

Algorithms that make the locally optimal choice at each stage in the hope of finding a global optimum.
Given a set of items, each with a weight and a value, and a knapsack with a maximum weight capacity, determine the maximum value that can be obtained by including fractions of items, such that the total weight does not exceed the knapsack capacity.

This code defines a function fractional_knapsack that solves the fractional knapsack problem using a greedy algorithm. It first sorts the items by their value-to-weight ratio in descending order. Then, it iterates through the items and takes the maximum possible fraction of each item, updating the total value and remaining capacity accordingly. The function returns the maximum value that can be obtained.

You can customize the items and capacity variables to test the code with different inputs.

```python
def fractional_knapsack(items, capacity):
    # Sort items by value-to-weight ratio in descending order
    items.sort(key=lambda x: x['value'] / x['weight'], reverse=True)

    total_value = 0
    for item in items:
        if capacity == 0:
            break

        weight_to_take = min(item['weight'], capacity)
        fraction = weight_to_take / item['weight']
        total_value += item['value'] * fraction
        capacity -= weight_to_take

    return total_value

if __name__ == "__main__":
    items = [
        {'value': 60, 'weight': 10},
        {'value': 100, 'weight': 20},
        {'value': 120, 'weight': 30},
    ]
    capacity = 50

    max_value = fractional_knapsack(items, capacity)
    print("The maximum value that can be obtained is:", max_value)
```

# Backtracking algorithm
Backtracking: A general algorithmic technique that incrementally builds candidates to the solutions, and abandons a candidate ("backtracks") as soon as it determines that the candidate cannot possibly be completed to a valid solution.

Place N chess queens on an N x N chessboard so that no two queens threaten each other (i.e., they should not be in the same row, column, or diagonal).

This code defines a function n_queens that solves the N-Queens problem using a backtracking algorithm. It uses helper functions is_safe and n_queens_util to check whether a queen can be placed safely in a given position and to recursively place queens on the board, respectively. The n_queens function returns the resulting chessboard with queens placed safely, or None if no solution is found.

You can customize the n variable to test the code with different board sizes.

```python
def is_safe(board, row, col, n):
    # Check the same column
    for i in range(row):
        if board[i][col] == 1:
            return False

    # Check upper diagonal on the left side
    for i, j in zip(range(row, -1, -1), range(col, -1, -1)):
        if board[i][j] == 1:
            return False

    # Check upper diagonal on the right side
    for i, j in zip(range(row, -1, -1), range(col, n)):
        if board[i][j] == 1:
            return False

    return True

def n_queens_util(board, row, n):
    if row == n:
        return True

    for col in range(n):
        if is_safe(board, row, col, n):
            board[row][col] = 1

            if n_queens_util(board, row + 1, n):
                return True

            board[row][col] = 0

    return False

def n_queens(n):
    board = [[0 for _ in range(n)] for _ in range(n)]

    if not n_queens_util(board, 0, n):
        return None

    return board

if __name__ == "__main__":
    n = 8
    result = n_queens(n)

    if result is not None:
        for row in result:
            print(row)
    else:
        print("No solution found.")
```

# Randomize algorithm
Randomized algorithms: Algorithms that use randomness as part of their logic.
Reservoir Sampling algorithm, which is used to select k items randomly from a list of items of unknown size. This algorithm is particularly useful when the size of the list is too large or expensive to bring into memory.

the reservoir_sampling function takes a stream (an iterator) and an integer k as input, and returns k items randomly selected from the stream using the Reservoir Sampling algorithm. The generate_stream function is a generator function that simulates a stream of items from 0 to n-1.

You can customize the k and n variables to test the code with different sample sizes and stream sizes.

```python
import random

def reservoir_sampling(stream, k):
    reservoir = []

    for i, item in enumerate(stream):
        if i < k:
            reservoir.append(item)
        else:
            random_index = random.randint(0, i)
            if random_index < k:
                reservoir[random_index] = item

    return reservoir

def generate_stream(n):
    for i in range(n):
        yield i

if __name__ == "__main__":
    k = 10
    n = 100

    stream = generate_stream(n)
    sampled_items = reservoir_sampling(stream, k)
    print("Sampled items:", sampled_items)
```

```python
import numpy as np

def evaluate(board):
    for row in range(3):
        if board[row, 0] == board[row, 1] == board[row, 2]:
            return board[row, 0]

    for col in range(3):
        if board[0, col] == board[1, col] == board[2, col]:
            return board[0, col]

    if board[0, 0] == board[1, 1] == board[2, 2]:
        return board[0, 0]

    if board[0, 2] == board[1, 1] == board[2, 0]:
        return board[0, 2]

    return 0

def minimax(board, depth, maximizing_player):
    evaluation = evaluate(board)
    if evaluation != 0 or depth == 0:
        return evaluation * (1 if maximizing_player else -1)

    if maximizing_player:
        max_eval = float('-inf')
        for row in range(3):
            for col in range(3):
                if board[row, col] == 0:
                    board[row, col] = 1
                    eval = minimax(board, depth - 1, False)
                    board[row, col] = 0
                    max_eval = max(max_eval, eval)
        return max_eval
    else:
        min_eval = float('inf')
        for row in range(3):
            for col in range(3):
                if board[row, col] == 0:
                    board[row, col] = -1
                    eval = minimax(board, depth - 1, True)
                    board[row, col] = 0
                    min_eval = min(min_eval, eval)
        return min_eval

def best_move(board):
    max_eval = float('-inf')
    move = (-1, -1)

    for row in range(3):
        for col in range(3):
            if board[row, col] == 0:
                board[row, col] = 1
                eval = minimax(board, 2, False)
                board[row, col] = 0

                if eval > max_eval:
                    max_eval = eval
                    move = (row, col)

    return move

if __name__ == "__main__":
    board = np.array([
        [ 1, -1,  0],
        [ 0,  1, -1],
        [ 0,  0,  0]
    ])

    optimal_move = best_move(board)
    print("Optimal move:", optimal_move)
```

The fft function works by recursively dividing the input sequence into even and odd indexed subsequences until the length is 1. It then combines the results of the even and odd subsequences using complex exponential functions.

You can test the code by modifying the length and values of the input sequence x. Make sure the length of x is a power of 2, as the Cooley-Tukey algorithm assumes this for simplicity.

```python
import numpy as np

def fft(x):
    N = len(x)
    if N <= 1:
        return x

    even = fft(x[0::2])
    odd = fft(x[1::2])

    T = [np.exp(-2j * np.pi * k / N) * odd[k] for k in range(N // 2)]
    return [even[k] + T[k] for k in range(N // 2)] + [even[k] - T[k] for k in range(N // 2)]

def ifft(x):
    N = len(x)
    x_conj = np.conjugate(x)
    y_conj = fft(x_conj)
    return np.conjugate(y_conj) / N

if __name__ == "__main__":
    x = np.random.rand(8)
    print("Original sequence:", x)

    X = fft(x)
    print("Fourier Transform (FFT):", X)

    x_reconstructed = ifft(X)
    print("Reconstructed sequence (IFFT):", x_reconstructed.real)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def reconstruct_signal(Y, freq_idx):
    Y_single_freq = np.zeros_like(Y)
    Y_single_freq[freq_idx] = Y[freq_idx]
    Y_single_freq[-freq_idx] = Y[-freq_idx]
    return np.fft.ifft(Y_single_freq)

# Generate a complex wave
N = 1024
t = np.linspace(0, 1, N)
f1, f2, f3 = 5, 15, 30
y = np.sin(2 * np.pi * f1 * t) + np.sin(2 * np.pi * f2 * t) + np.sin(2 * np.pi * f3 * t)

# Add noise
noise = np.random.normal(0, 1, N)
y_noisy = y + noise

# Compute FFT of the complex wave
Y = np.fft.fft(y_noisy)

# Find the magnitude and normalize
magnitude = np.abs(Y) / N

# Compute the frequencies
frequencies = np.fft.fftfreq(N, d=t[1] - t[0])

# Filter the signal by removing low magnitude components
threshold = 0.1
significant_freq_indices = np.where(magnitude > threshold)[0]

# Reconstruct each periodic signal
reconstructed_signals = [reconstruct_signal(Y, freq_idx) for freq_idx in significant_freq_indices]

# Plot each periodic signal
for i, reconstructed_signal in enumerate(reconstructed_signals, 1):
    plt.subplot(len(reconstructed_signals), 1, i)
    plt.plot(t, reconstructed_signal.real)
    plt.title(f"Periodic Signal {i}")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
def func(x): # Define the function to be integrated
    return x**2
a, b, N= 0, 2, 100000 # Limit 0~2 and number of random points 100000 
c = b**2
x = np.random.uniform(a, b, N) # Generate random x and y values between the limits of integration
y = np.random.uniform(a, c, N)
f = func(x) # Evaluate the function at the random x values
below = np.sum(y < f) # Count the number of points below the function
integral = b*c * (below/N) # Calculate the area under the function
print("Monte Carlo Estimate of the Integral: ", integral, 'Analytic Integral:', 1/3*b**3) # Print the result
plt.scatter(x, y, alpha=0.1, s=3, c='g', label='Random Points') # Plot the results
plt.scatter(x, f, c='b', label='Function')
plt.fill_between(np.linspace(a,b,num=20), 0, func(np.linspace(a,b,num=20)), alpha=0.8, color='r', label='Area Under Function')
plt.legend()
plt.xlim(0,b)
plt.ylim(0,c)
plt.show()
```

# Random Walk

```python
import matplotlib.pyplot as plt
import numpy as np
num_walks = 3 # Set up the parameters
num_steps = 500
step_size = 1
start_positions = [[0, 0, 0], [1, 1, 1], [2, -1, 3]] # Define the starting positions for each walker
walks = [] # Generate the random walks
for i in range(num_walks):
    x, y, z = start_positions[i]
    walk = [[x], [y], [z]]
    for j in range(num_steps):
        step = step_size * np.random.normal(0, 4, 1).astype(int)[0]
        xyz=np.random.randint(0,3)
        if xyz == 0:
            x+=step
        elif xyz == 1:
            y+=step
        else :
            z+=step
        walk[0].append(x)
        walk[1].append(y)
        walk[2].append(z)
    walks.append(walk)
fig = plt.figure(figsize=(8, 8)) # Plot the random walks
ax = fig.add_subplot(111, projection='3d')
for i in range(num_walks):
    x, y, z = walks[i]
    ax.plot(x, y, z)
plt.show()
```

# Black-Scholes Equation

```python
import numpy as np
from scipy.stats import norm
def black_scholes(S, K, T, r, sigma, option_type):
    d1 = (np.log(S/K) + (r + 0.5*sigma**2)*T) / (sigma*np.sqrt(T))
    d2 = d1 - sigma*np.sqrt(T)
    if option_type == 'call':
        price = S*norm.cdf(d1) - K*np.exp(-r*T)*norm.cdf(d2)
    elif option_type == 'put':
        price = K*np.exp(-r*T)*norm.cdf(-d2) - S*norm.cdf(-d1)
    return price
S = 100   # current stock price
K = 110   # strike price
T = 1     # time to maturity (in years)
r = 0.05  # risk-free interest rate
sigma = 0.2  # volatility
option_type = 'call'  # type of option ('call' or 'put')
price = black_scholes(S, K, T, r, sigma, option_type)
print(price)  # print the option price
```

```python
import cv2
import numpy as np
import svgwrite
import svgpathtools
def vectorize_image(image_path, output_path):
    # Read the image
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Threshold the image
    _, binary_image = cv2.threshold(image, 128, 255, cv2.THRESH_BINARY_INV)

    # Find contours
    contours, _ = cv2.findContours(binary_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Create an SVG canvas
    dwg = svgwrite.Drawing(output_path, profile='tiny', size=(image.shape[1], image.shape[0]))

    # Add contours to the SVG canvas
    for contour in contours:
        points = [tuple(map(float, point[0])) for point in contour]
        polyline = dwg.polyline(points, stroke='black', fill='none', stroke_width=1)
        dwg.add(polyline)

    # Save the SVG file
    dwg.save()

image_path = 'img/left.png'
output_path = 'img/output.svg'
vectorize_image(image_path, output_path)
paths, attributes = svgpathtools.svg2paths('img/output.svg')
```

## 6.6 High Dynamic Range Image (HDR)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
def readImagesAndTimes():
    filenames = ["img/img_0.033.jpg", "img/img_0.25.jpg", "img/img_2.5.jpg", "img/img_15.jpg"]
    times = np.array([ 1/30.0, 0.25, 2.5, 15.0 ], dtype=np.float32)
    images = []
    for filename in filenames:
        im = cv2.imread(filename)
        im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
        images.append(im)
    return images, times
# Read images and exposure times
images, times = readImagesAndTimes()

# Align Images
alignMTB = cv2.createAlignMTB()
alignMTB.process(images, images)
# Merge images into an HDR linear image
calibrateDebevec = cv2.createCalibrateDebevec()
mergeDebevec = cv2.createMergeDebevec()
responseDebevec = calibrateDebevec.process(images, times)
hdrDebevec = mergeDebevec.process(images, times, responseDebevec)
# Tonemap using Drago's method to obtain 24-bit color image
tonemapDrago = cv2.createTonemapDrago(1.0, 0.7)
ldrDrago = tonemapDrago.process(hdrDebevec)
ldrDrago = 3 * ldrDrago
plt.figure(figsize=(20,10)); plt.imshow(np.clip(ldrDrago,0,1)); plt.axis('off');
cv2.imwrite("img/ldr-Drago.jpg", ldrDrago * 255)
print("img.saved_Drago_Tonemap.jpg")
# Tonemap using Reinhard's method to obtain 24-bit color image
tonemapReinhard = cv2.createTonemapReinhard(1.5, 0,0,0)
ldrReinhard = tonemapReinhard.process(hdrDebevec)
plt.figure(figsize=(20,10)); plt.imshow(np.clip(ldrReinhard,0,1)); plt.axis('off');
cv2.imwrite("img/ldr-Reinhard.jpg", ldrReinhard * 255)
print("saved_Reinhard_Tonemap.jpg")
# Tonemap using Mantiuk's method to obtain 24-bit color image
tonemapMantiuk = cv2.createTonemapMantiuk(2.2,0.85, 1.2)
ldrMantiuk = tonemapMantiuk.process(hdrDebevec)
ldrMantiuk = 3 * ldrMantiuk
plt.figure(figsize=(20,10)); plt.imshow(np.clip(ldrMantiuk,0,1)); plt.axis('off');
cv2.imwrite("img/ldr-Mantiuk.jpg", ldrMantiuk * 255)
print("img/saved_Mantiuk_Tonemap.jpg")
```

## 6.7 Image alignment
1. A **Homography** transforms a square to arbitrary quad.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
def extract_features(image, method, features):
    if method == 'SURF':
        surf = cv2.xfeatures2d.SURF_create(features) #patent
        kp, des = surf.detectAndCompute(image, None)
    elif method == 'SIFT':
        sift = cv2.SIFT_create(features)
        kp, des = sift.detectAndCompute(image, None)
    elif method == 'ORB':
        orb = cv2.ORB_create(features)
        kp, des = orb.detectAndCompute(image, None)
    return kp, des

def match_features(des1, des2, method):
    if method == 'BF':
        bf = cv2.BFMatcher(crossCheck=True)
        matches = bf.knnMatch(des1, des2, k=2)
        matches = sorted(matches, key=lambda x: x.distance)
    elif method == 'SIFT':
        bf = cv2.BFMatcher()
        raw_matches = bf.knnMatch(des1, des2, k=2)
        matches = []
        for m, n in raw_matches:
            if m.distance < 0.55 * n.distance:  # Ratio test
                matches.append(m)
        matches = sorted(matches, key=lambda x: x.distance)
    elif method == 'ORB':
        matcher = cv2.DescriptorMatcher_create(cv2.DESCRIPTOR_MATCHER_BRUTEFORCE_HAMMING)
        matches = matcher.match(des1, des2, None)
        numGoodMatches = int(len(matches) * 0.01)
        matches = matches[:numGoodMatches]
    else:
        FLANN_INDEX_KDTREE = 1
        index_params = dict(algorithm=FLANN_INDEX_KDTREE, trees=5)
        search_params = dict(checks=50)
        flann = cv2.FlannBasedMatcher(index_params, search_params)
        matches = flann.knnMatch(des1, des2, k=2)
    return matches

def visualize_matches(image1, kp1, image2, kp2, matches):
    image_matches = cv2.drawMatches(image1, kp1, image2, kp2, matches, None, flags=2)
    plt.figure(figsize=(16, 6), dpi=100)
    plt.imshow(image_matches)

# Read reference image
refFilename = "img/form.jpg"
im1 = cv2.imread(refFilename, cv2.IMREAD_COLOR)
im1 = cv2.cvtColor(im1, cv2.COLOR_BGR2RGB)
im1_gray = cv2.cvtColor(im1, cv2.COLOR_BGR2GRAY)

# Read image to be aligned
imFilename = "img/scanned-form.jpg"
im2 = cv2.imread(imFilename, cv2.IMREAD_COLOR)
im2 = cv2.cvtColor(im2, cv2.COLOR_BGR2RGB)
im2_gray = cv2.cvtColor(im2, cv2.COLOR_BGR2GRAY)

# Detect ORB features and compute descriptors.
MAX_NUM_FEATURES = 500
key1, des1 = extract_features(im1_gray, 'SIFT', MAX_NUM_FEATURES)
key2, des2 = extract_features(im2_gray, 'SIFT', MAX_NUM_FEATURES)

# Match features.
matches = match_features(des1, des2, 'SIFT')

# Visualize matches
visualize_matches(im1_gray, key1, im2_gray, key2, matches)
plt.show()

# Extract location of good matches
points1 = np.zeros((len(matches), 2), dtype=np.float32)
points2 = np.zeros((len(matches), 2), dtype=np.float32)

for i, match in enumerate(matches):
    points1[i, :] = key1[match.queryIdx].pt
    points2[i, :] = key2[match.trainIdx].pt
  
# Find homography
h, mask = cv2.findHomography(points2, points1, cv2.RANSAC, 5.0)

# Use homography to warp image
height, width, channels = im1.shape
im2_reg = cv2.warpPerspective(im2, h, (width, height))
# Display results 
plt.figure(figsize=[20,10]); 
plt.subplot(121); plt.imshow(im1); plt.axis('off'); plt.title("Original Form");
plt.subplot(122); plt.imshow(im2_reg); plt.axis('off'); plt.title("Scanned Form");
```

## 6.8 Extract meta data of image

```python
import folium
from PIL import Image
from PIL.ExifTags import TAGS

def get_exif_data(image_path):
    image = Image.open(image_path)
    exif_data = {}
    if image._getexif() is not None:
        for tag, value in image._getexif().items():
            if tag in TAGS:
                exif_data[TAGS[tag]] = value
    return exif_data

def convert_to_decimal(gps_data):
    degrees, minutes, seconds = gps_data
    decimal = degrees + minutes / 60 + seconds / 3600
    return decimal

def get_gps_decimal(gps_info):
    latitude = convert_to_decimal(gps_info[2])
    longitude = convert_to_decimal(gps_info[4])
    if gps_info[1] == "S":
        latitude = -latitude
    if gps_info[3] == "W":
        longitude = -longitude
    return latitude, longitude
"""def create_map(latitude, longitude, date_info):
    map = folium.Map(location=[latitude, longitude], zoom_start=16)
    folium.Marker([latitude, longitude], popup=f"Date: {date_info}").add_to(map)
    return map"""
def create_map_with_markers(image_paths):
    # Initialize an empty map centered at the first image's location
    first_image_exif_data = get_exif_data(image_paths[0])
    first_image_gps_info = first_image_exif_data["GPSInfo"]
    first_image_latitude, first_image_longitude = get_gps_decimal(first_image_gps_info)
    map = folium.Map(location=[first_image_latitude, first_image_longitude], zoom_start=5)

    # Loop through the image paths and add markers to the map
    for img_path in image_paths:
        exif_data = get_exif_data(img_path)

        if "GPSInfo" in exif_data:
            gps_info = exif_data["GPSInfo"]
            latitude, longitude = get_gps_decimal(gps_info)

            # Get the date information from the EXIF data
            date_info = exif_data.get("DateTimeOriginal", "Date not found")

            # Add the marker with the date popup to the map
            folium.Marker([latitude, longitude], popup=f"Date: {date_info}").add_to(map)
        else:
            print(f"No GPS information found in the image: {img_path}")

    return map

image_path =["img/IMG_0983.JPG", 'img/IMG_0926.JPG']

map_with_markers = create_map_with_markers(image_path)
map_with_markers.save("outputs/Image_map.html")
```
